```python
from kittentts import KittenTTS
m = KittenTTS("KittenML/kitten-tts-nano-0.1")
```

```python
from IPython.display import Audio
```

```python
audio = m.generate("A breathtaking view of the Grand Canyon at sunrise, layers of red rock glowing in the golden light, the Colorado River snaking through the canyon below. Photorealistic landscape, high detail, wide-angle shot.", voice='expr-voice-2-m')

# available_voices : [  'expr-voice-2-m', 'expr-voice-2-f', 'expr-voice-3-m', 'expr-voice-3-f',  'expr-voice-4-m', 'expr-voice-4-f', 'expr-voice-5-m', 'expr-voice-5-f' ]

# Save the audio
import soundfile as sf
sf.write('output.wav', audio, 24000)

Audio('output.wav')
```
